```python
import pandas as pd
```

```python
df=pd.read_csv("FDI data.csv")
```

```python
df_long = df.melt(id_vars=['Sector'], var_name='Year', value_name='FDI')
```

```python
df_long['Year'] = pd.to_datetime(df_long['Year'].str[:4] + '-04-01')
```

```python
yearly_fdi = df_long.groupby('Year')['FDI'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=yearly_fdi, x='Year', y='FDI', marker='o')
plt.title('Year-wise Total FDI in India (2000-01 to 2016-17)')
plt.xlabel('Year')
plt.ylabel('Total FDI (in million USD)')
plt.grid(True)
plt.show()
```

```python
sector_fdi = df_long.groupby('Sector')['FDI'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=sector_fdi, x='Sector', y='FDI')
plt.title('Sector-wise Total FDI in India (2000-01 to 2016-17)')
plt.xlabel('Sector')
plt.ylabel('Total FDI (in million USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
yearly_fdi['YoY_Growth'] = yearly_fdi['FDI'].pct_change() * 100
```
